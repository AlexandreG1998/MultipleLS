```python
def find_cointegrate (var_dependente = None, var_independente = None, siglevel = 5, 
                      cut = 0.5):
    
    split = int(len(var_dependente) * cut)
    
    treino_x = var_independente[:split]
    treino_y = var_dependente[:split]
    
    model = LinearRegression().fit(treino_x.values,treino_y.values.ravel())
    prev  = model.predict(treino_x.values)
    resid = pd.DataFrame(treino_y.values.ravel()) - pd.DataFrame(prev)
    sig = adfuller(resid.values.ravel())[1] * 100
    
    return(sig)
```

```python
def get_sort (size = 5, len_ = 10, comb_total = 10000):
    
    comb = []
    
    while(len(comb) != comb_total):
        comb_ = np.random.randint(low = 0, high = len_, size = size).tolist()
        if((comb_ in comb) == False):
            comb_final = []
            for i in range(len(comb_)):
                if((comb_[i] in comb_final)== False):
                    comb_final.append(comb_[i])
            if(len(comb_final) == size):
                comb.append(comb_final)
    return(comb)
```

```python
def backtest (var_dependente = None, var_independente = None, siglevel = 5, 
                      cut = 0.5, Y = 5, acoes = 1000, bs_level = 70, bi_level = 30):
    
    
    var_dependente   = var_dependente.reset_index().iloc[:,1:]
    var_independente = var_independente.reset_index().iloc[:,1:]
    
    split = int(len(var_dependente) * cut)
    
    treino_x = var_independente[:split]
    treino_y = var_dependente[:split]
    
    model = LinearRegression().fit(treino_x.values,treino_y.values.ravel())
    prev  = model.predict(treino_x.values)
    resid = pd.DataFrame(treino_y.values.ravel()) - pd.DataFrame(prev)
    
    bs = np.percentile(resid, q = bs_level)
    bi = np.percentile(resid, q = bi_level)
    
    
    data = pd.concat((var_dependente, var_independente), axis = 1)
    data['Previsto'] = model.predict(data.iloc[:,1:].values)
    data['Residuos'] = data[data.columns[0]] - data['Previsto']
    data['Y']   = ((data[data.columns[0]].shift(-Y)/data[data.columns[0]]) - 1) * 100
    data['Yfin']   = ((data[data.columns[0]].shift(-Y)-data[data.columns[0]]))
    data['Sinal']    = np.where((data['Residuos'] > bs),-1, np.where((data['Residuos'] < bi),1,0))
    data = data.dropna()
    data = data[split:].reset_index().iloc[:,1:]
    data['Sinal_adj'] = np.where((data['Sinal'].shift() == 0) &
                                 (data['Sinal'].shift(2) == 0) &
                                 (data['Sinal'].shift(3) == 0) &
                                 (data['Sinal'].shift(4) == 0), data['Sinal'], 0)
    data['Pontos']     = data['Sinal_adj'] * data['Y']
    data['Financeiro'] = data['Sinal_adj'] * (data['Yfin'] * acoes)
    if(len(data[data['Pontos'] != 0]) > 5):
        sharpe = np.mean(data['Pontos'])/np.std(data['Pontos'])
        rrp     = abs(data[data['Pontos'] > 0]['Pontos'].sum() / data[data['Pontos'] < 0]['Pontos'].sum()+ 0.01)
        rrf     = abs(data[data['Financeiro'] > 0]['Financeiro'].sum() / data[data['Financeiro'] < 0]['Financeiro'].sum() + 0.01)
        num_operacoes = len(data[data['Pontos'] != 0])
        tx_acerto     = (len(data[data['Pontos'] > 0])/len(data[data['Pontos'] != 0])) * 100
        dados_backtest = [sharpe,rrp,rrf, num_operacoes, tx_acerto]
        return([dados_backtest, data['Pontos'].values, data['Financeiro'].values])
    else:
        return(['erro', 'erro', 'erro'])
```

```python
combinacao_backtest       = []
dados_backtest_compilados = []
pontos_backtest           = []
financeiro_backtest       = []

for i in range(len(selecao)):
    
    combinacao_atual = selecao.iloc[:,0:1][i: i + 1].values.ravel()[0].split("_")
    
    data = pd.concat((conjunto_ativos_ls[combinacao_atual[1]],
                      conjunto_ativos_ls[combinacao_atual[2]],
                      conjunto_ativos_ls[combinacao_atual[3]],
                      conjunto_ativos_ls[combinacao_atual[4]],
                      conjunto_ativos_ls[combinacao_atual[5]]), axis = 1)
    
    retorno_backtest = backtest (var_dependente = data.iloc[:,0:1], var_independente = data.iloc[:,1:], siglevel = 5, 
                      cut = 0.5, Y = 5, acoes = 1000, bs_level = 70, bi_level = 30)
    combinacao_backtest.append(selecao.iloc[:,0:1][i: i + 1].values.ravel()[0])
    dados_backtest_compilados.append(retorno_backtest[0])
    pontos_backtest.append(retorno_backtest[1])
    financeiro_backtest.append(retorno_backtest[2])
```
